### Q1: Implement Selection Sort on a list of integers.

Write a function `selection_sort(arr)` that implements selection sort.
**Example**:
Input: [64, 25, 12, 22, 11]
Output: [11, 12, 22, 25, 64]

```python
def selection_sort(arr):
    n = len(arr)
    for i in range(n):
        min_idx = i
        for j in range(i+1, n):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]

# Example usage
arr = [64, 25, 12, 22, 11]
selection_sort(arr)
print(arr)
```

### Q2: Count Swaps in Bubble Sort

Modify the bubble sort algorithm to count and return the total number of swaps made during the sorting.
**Example**:
Input: [3, 2, 1]
Output: 3

```python
def bubble_sort_count_swaps(arr):
    n = len(arr)
    swaps = 0
    for i in range(n):
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
                swaps += 1
    return swaps

# Example usage
arr = [3, 2, 1]
swaps = bubble_sort_count_swaps(arr)
print('Swaps:', swaps)
```

### Q3: Find the Kth Smallest Element Using Bubble Sort

Given a list of integers and an integer k, use bubble sort to find the kth smallest element.
Sort only till kth pass.
**Example**:
Input: arr = [7, 10, 4, 3, 20, 15], k = 3
Output: 7

```python
def kth_smallest_bubble(arr, k):
    n = len(arr)
    for i in range(k):
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    return arr[k-1]

# Example usage
arr = [7, 10, 4, 3, 20, 15]
k = 3
print('Kth smallest:', kth_smallest_bubble(arr, k))
```

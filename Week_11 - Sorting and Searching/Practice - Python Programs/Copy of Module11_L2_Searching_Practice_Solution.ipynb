## 1. Search in a Rotated Sorted Array

You are given a list that was sorted but then rotated. Write a function to search for a target element and return its index. If not found, return -1.

Example:
Input: nums = [6,7,8,1,2,3,4,5], target = 3
Output: 5

```python
def search_rotated_array(nums, target):
    low, high = 0, len(nums) - 1
    while low <= high:
        mid = (low + high) // 2
        if nums[mid] == target:
            return mid
        if nums[low] <= nums[mid]:
            if nums[low] <= target < nums[mid]:
                high = mid - 1
            else:
                low = mid + 1
        else:
            if nums[mid] < target <= nums[high]:
                low = mid + 1
            else:
                high = mid - 1
    return -1
```

## 2. Find the First and Last Occurrence of a Number

Write a function that takes a sorted list and a target number, and returns the first and last index where the number appears. Return [-1, -1] if the number is not found.

Example:
Input: nums = [2,4,4,4,6,7], target = 4
Output: [1, 3]

```python
def find_first_last(nums, target):
    def binary_search_left():
        low, high = 0, len(nums) - 1
        left = -1
        while low <= high:
            mid = (low + high) // 2
            if nums[mid] < target:
                low = mid + 1
            else:
                if nums[mid] == target:
                    left = mid
                high = mid - 1
        return left

    def binary_search_right():
        low, high = 0, len(nums) - 1
        right = -1
        while low <= high:
            mid = (low + high) // 2
            if nums[mid] > target:
                high = mid - 1
            else:
                if nums[mid] == target:
                    right = mid
                low = mid + 1
        return right

    return [binary_search_left(), binary_search_right()]
```

## 3. Find Peak Element

A peak element is one that is greater than its neighbors. Find the index of any peak element in the list.

Example:
Input: nums = [1, 3, 20, 4, 1, 0]
Output: 2

```python
def find_peak(nums):
    low, high = 0, len(nums) - 1
    while low < high:
        mid = (low + high) // 2
        if nums[mid] > nums[mid + 1]:
            high = mid
        else:
            low = mid + 1
    return low
```

## 4. Search Insert Position

Given a sorted list and a target value, return the index if the target is found. If not, return the index where it would be inserted. Don't use Linear Search.

Example:
Input: nums = [1,3,5,6], target = 2
Output: 1

```python
def search_insert(nums, target):
    low, high = 0, len(nums)
    while low < high:
        mid = (low + high) // 2
        if nums[mid] < target:
            low = mid + 1
        else:
            high = mid
    return low
```

# W11_L1_V5 - Selection and Bubble Sort Implementation

```python
import numpy as np
```

```python
# Bubble Sort
```

```python
def bubble_sort_np(arr: np.array) -> np.ndarray:
    """
    In-place bubble sort on a 1-D NumPy array.
    Prints the array after each outer pass.
    """
    n = arr.size
    for i in range(n-1):
        for j in range(n-1-i):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
        print(f"Bubble pass {i + 1} -> {arr}")
    return arr
```

```python
# Selction Sort
```

```python
def selection_sort_np(arr: np.array) -> np.ndarray:
    """
    In-place selection sort on a 1-D NumPy array.
    Prints the array after each outer pass.
    """
    n = arr.size
    for i in range(n-1):
        min_idx = i
        for j in range(i + 1, n):
            if arr[j] < arr[min_idx]:
                min_idx = j
        if min_idx != i:
            arr[i], arr[min_idx] = arr[min_idx], arr[i]
        print(f"Selection pass {i + 1} -> {arr}")
    return arr
```

```python
# Demo
```

```python
if __name__ == "__main__":
    original = np.array([44, 42, 19, 35, 10, 27, 33, 14])
    print("Original: ", original)
    
    print("\n Bubble Sort Trace")
    bubble_sorted = bubble_sort_np(original.copy())
    
    print("\n Selection Sort Trace")
    selection_sorted = selection_sort_np(original.copy())
    
    print("\n Final Sorted Array: ", selection_sorted)
    
    
```

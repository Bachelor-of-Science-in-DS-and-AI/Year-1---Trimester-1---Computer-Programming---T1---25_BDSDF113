# W11_L2_V2 - Linear Search

```python
import numpy as np
```

```python
# Linear Search
```

```python
def linear_search_np(arr: np.ndarray, target):
    n = arr.size
    for i in range(n):
        if arr[i] == target:
            return i
    return -1

if __name__ == "__main__":
    data = np.array([10, 23, 45, 23, 17])
    x = 23
    pos = linear_search_np(data, x)
    
    if pos != -1:
        print(f"{x} found at index {pos}")
    else:
        print(f"{x} not found in array")        
```

```python
# Binary Search(Iterative)
```

```python
import numpy as np
```

```python
def binary_search_np(arr: np.ndarray, target: int) -> int:
    """
    Iterative binary search on a **sorted** 1-D NumPy array.
    Returns the index of target 0r -1 if not found
    """
    left, right = 0, arr.size - 1
    while left <= right:
        mid = (left + right) // 2
        print(f"Searching... left = {left}, right = {right}, mid = {mid}, arr[mid] = {arr[mid]}")
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else: 
            right = mid - 1
    return -1
```

```python
# Demo
```

```python
original = np.array([44, 42, 19, 35, 10, 27, 33, 14])
print("Original: ", original)
sortedArr = np.sort(original)
print("Sorted: ", sortedArr)

print("\n Binary Search Demo")
target = 35
idx = binary_search_np(sortedArr, target)
print(f"Index of {target}: " , idx)
```

Q1 Using the Python math library, write a function that calculates the distance between two points on a circular path (arc length), given:

The radius r of the circle.

The central angle theta in degrees between the two points on the circle.

Then, using this function, compute the arc length for a circle of radius 10 units and a central angle of 75 degrees.

Use the math library for any needed conversions or calculations.

```python
import math

def arc_length(radius, angle_degrees):
    # Convert degrees to radians
    angle_radians = math.radians(angle_degrees)

    # Arc length formula: s = r * θ (θ in radians)
    length = radius * angle_radians
    return length

# Example usage
radius = 10
angle = 75

result = arc_length(radius, angle)
print(f"The arc length is: {result:.2f} units")
```

Q2 Create a program that asks the user to enter their date of birth (in YYYY-MM-DD format). Print how many days are left until their next birthday.

Hint: You will need to compare today's date with the next occurrence of their birthday using datetime.date.

```python
import datetime

dob_input = input("Enter your date of birth (YYYY-MM-DD): ")
dob = datetime.datetime.strptime(dob_input, "%Y-%m-%d").date()
today = datetime.date.today()

# Calculate the next birthday
next_birthday = dob.replace(year=today.year)
if next_birthday < today:
    next_birthday = next_birthday.replace(year=today.year + 1)

days_remaining = (next_birthday - today).days
print(f"Your next birthday is in {days_remaining} day(s).")
```

Q3 Write a Python program that calculates the area of a right triangle using the Pythagorean theorem. The user should input the lengths of the two perpendicular sides (a and b), and your program should compute the hypotenuse and then return the area using:

area = 0.5 * a * b

hypotenuse = sqrt(a² + b²)
Use the math library for any needed conversions or calculations.

```python
import math

a = float(input("Enter side a: "))
b = float(input("Enter side b: "))

hypotenuse = math.sqrt(a**2 + b**2)
area = 0.5 * a * b

print(f"Hypotenuse: {hypotenuse}")
print(f"Area of the triangle: {area}")
```

Q4 Simulate a simple dice game between two players. Each player rolls a 6-sided die (use random.randint). Repeat the game 5 times and keep track of wins for each player.

At the end, print the winner or declare a tie based on the total number of wins.

```python
import random

player1_wins = 0
player2_wins = 0

for i in range(5):
    roll1 = random.randint(1, 6)
    roll2 = random.randint(1, 6)
    print(f"Round {i+1}: Player 1 rolled {roll1}, Player 2 rolled {roll2}")

    if roll1 > roll2:
        player1_wins += 1
    elif roll2 > roll1:
        player2_wins += 1
    else:
        print("It's a tie this round!")

print(f"Player 1 wins: {player1_wins}")
print(f"Player 2 wins: {player2_wins}")

if player1_wins > player2_wins:
    print("Player 1 is the overall winner!")
elif player2_wins > player1_wins:
    print("Player 2 is the overall winner!")
else:
    print("The game is a tie!")
```
